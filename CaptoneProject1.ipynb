```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Download the datasets
# Note: Replace these URLs with direct links to the CSV files if needed.

# Read the data
male_data = pd.read_csv("G2.csv")
female_data = pd.read_csv("G1.csv")

# 2. Convert to numpy matrices
male = male_data[['weight', 'standing_height', 'upper_arm_length', 'upper_leg_length', 'arm_circumference', 'hip_circumference', 'waist_circumference']].to_numpy()
female = female_data[['weight', 'standing_height', 'upper_arm_length', 'upper_leg_length', 'arm_circumference', 'hip_circumference', 'waist_circumference']].to_numpy()

# 3. Plot histograms
plt.figure(figsize=(12, 8))

# Female weights histogram
plt.subplot(2, 1, 1)
plt.hist(female[:, 0], bins=20, color='pink', alpha=0.7)
plt.title('Female Weights Histogram')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.xlim(40, 150)  # Example limits, adjust based on data

# Male weights histogram
plt.subplot(2, 1, 2)
plt.hist(male[:, 0], bins=20, color='blue', alpha=0.7)
plt.title('Male Weights Histogram')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.xlim(40, 150)  # Same limits for comparison

plt.tight_layout()
plt.show()

# 4. Box-and-whisker plot
plt.figure(figsize=(8, 6))
plt.boxplot([female[:, 0], male[:, 0]], labels=['Female', 'Male'])
plt.title('Weight Comparison')
plt.ylabel('Weight (kg)')
plt.show()

# 5. Basic numerical aggregates
female_weights = female[:, 0]
male_weights = male[:, 0]

female_stats = {
    'mean': np.mean(female_weights),
    'std': np.std(female_weights),
    'median': np.median(female_weights),
    'min': np.min(female_weights),
    'max': np.max(female_weights)
}

male_stats = {
    'mean': np.mean(male_weights),
    'std': np.std(male_weights),
    'median': np.median(male_weights),
    'min': np.min(male_weights),
    'max': np.max(male_weights)
}

print("Female Weights Statistics:", female_stats)
print("Male Weights Statistics:", male_stats)

# 6. Add BMI to female matrix
bmi_female = female_weights / (female[:, 1] / 100) ** 2
female = np.column_stack((female, bmi_female))

# 7. Standardize the female dataset
zfemale = (female - np.mean(female, axis=0)) / np.std(female, axis=0)

# 8. Scatterplot matrix and correlation coefficients
sns.pairplot(pd.DataFrame(zfemale, columns=['weight', 'standing_height', 'upper_arm_length', 'upper_leg_length', 'arm_circumference', 'hip_circumference', 'waist_circumference', 'BMI']))
plt.show()

# Calculate Pearson's and Spearman's correlations
correlations_pearson = np.corrcoef(zfemale, rowvar=False)
correlations_spearman = pd.DataFrame(zfemale).corr(method='spearman')

print("Pearson Correlation Coefficients:\n", correlations_pearson)
print("Spearman Correlation Coefficients:\n", correlations_spearman)

# 9. Waist circumference ratios
male_waist_height_ratio = male[:, 6] / male[:, 1]
female_waist_height_ratio = female[:, 6] / female[:, 1]
male_waist_hip_ratio = male[:, 6] / male[:, 5]
female_waist_hip_ratio = female[:, 6] / female[:, 5]

male = np.column_stack((male, male_waist_height_ratio, male_waist_hip_ratio))
female = np.column_stack((female, female_waist_height_ratio, female_waist_hip_ratio))

# 10. Box-and-whisker plot for ratios
plt.figure(figsize=(12, 6))
plt.boxplot([female_waist_height_ratio, female_waist_hip_ratio, male_waist_height_ratio, male_waist_hip_ratio],
            labels=['Female Waist/Height', 'Female Waist/Hip', 'Male Waist/Height', 'Male Waist/Hip'])
plt.title('Waist Circumference Ratios Comparison')
plt.ylabel('Ratio')
plt.show()

# 11. Discuss advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio

# 12. Print standardized body measurements for lowest and highest BMI
lowest_bmi_indices = np.argsort(zfemale[:, 7])[:5]
highest_bmi_indices = np.argsort(zfemale[:, 7])[-5:]

print("Lowest BMI Measurements:\n", zfemale[lowest_bmi_indices])
print("Highest BMI Measurements:\n", zfemale[highest_bmi_indices])
```
